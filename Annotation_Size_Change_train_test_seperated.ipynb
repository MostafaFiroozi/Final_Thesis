```python
import os
import glob
```

```python
rout_path=r"C:\Users\39351\Desktop\Images1"
```

```python
# import element tree
import xml.etree.ElementTree as ET 
path=os.path.join(rout_path,'train','*.xml')
#print(path)
for file in glob.glob(path):
        #print(file)
        #import xml file
        tree = ET.parse(file)
        root = tree.getroot()
        elems = tree.findall(".//xmin")
        print(elems)
        for elem in elems:
            print(elem.text)
            elem.text=str(int(elem.text)*2)
            print(elem.text)
        elems = tree.findall(".//ymin")
        print(elems)
        for elem in elems:
            print(elem.text)
            elem.text=str(int(elem.text)*2)
            print(elem.text)
        elems = tree.findall(".//xmax")
        print(elems)
        for elem in elems:
            print(elem.text)
            elem.text=str(int(elem.text)*2)
            print(elem.text)
        elems = tree.findall(".//ymax")
        print(elems)
        for elem in elems:
            print(elem.text)
            elem.text=str(int(elem.text)*2)
            print(elem.text)
        out_name=os.path.basename(file[0:-4])+'.xml'
        out_path=os.path.join(rout_path,'Train_Test_800','train',out_name)
        print(out_path)
        tree.write(out_path)
```
