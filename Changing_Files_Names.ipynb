```python
from PIL import Image
import glob
import os
import pathlib
```

```python
path= os.path.join(r"C:\Users\39351\Desktop\Images\newSetAlum_color\resized_images_400",'*.*')
files_dir=r"C:\Users\39351\Desktop\Images\newSetAlum_color\resized_images_400"
for file in glob.glob(path):
        file_name=name=file[len(path)-3:-4]
        file_format=pathlib.Path(file).suffix
        new_name='newSetAlum_color'+name+file_format
        new_file=os.path.join(files_dir,'newSetAlum_color'+file_name+file_format)
        print(new_file)
        os.rename(file, new_file)
```
